```python
MLS_dict = {'comp': 'MLS',
             'link': 'https://oddspedia.com/br/futebol/eua/mls',
             'semanas' : '30'
            }

BRA_A_dict = {'comp': 'BRA_A',
             'link': 'https://oddspedia.com/br/futebol/brasil/brasileirao-serie-a',
             'semanas' : '30'
              } 

BRA_B_dict = {'comp': 'BRA_B',
             'link': 'https://oddspedia.com/br/futebol/brasil/brasileirao-serie-b',
             'semanas' : '31'
              } 

SUE_A_dict =  {'comp': 'SUE_A',
             'link': 'https://oddspedia.com/br/futebol/suecia/liga-allsvenskan',
             'semanas' : '29'
              }  

CHIN_A_dict =  {'comp': 'CHIN_A',
             'link': 'https://oddspedia.com/br/futebol/china/super-liga-chinesa',
             'semanas' : '24'
              }                         

JAP_A_dict =  {'comp': 'JAP_A',
             'link': 'https://oddspedia.com/br/futebol/japao/liga-j1',
             'semanas' : '34'
              } 


NOR_A_dict =  {'comp': 'NOR_A',
             'link': 'https://oddspedia.com/br/futebol/noruega/elitserien',
             'semanas' : '29'
              }

BEL_A_dict = {'comp': 'BEL_A',
             'link': 'https://oddspedia.com/br/futebol/belgica/primeira-divisao-a',
             'semanas' : '29'
             }

USL_A_dict = {'comp': 'USL_A',
             'link': 'https://oddspedia.com/br/futebol/eua/liga-unida-de-futebol',
             'semanas' : '32'
             }             


ING_B_dict = {'comp': 'ING_B',
             'link': 'https://oddspedia.com/br/futebol/inglaterra/championship',
             'semanas' : '36'
             } 


ALE_B_dict = {'comp': 'ALE_B',
             'link': 'https://oddspedia.com/br/futebol/alemanha/2a-bundesliga',
             'semanas' : '34'
             }             
lista_dicts = [
                BRA_A_dict, 
                #BRA_B_dict,
                SUE_A_dict,
                #BEL_A_dict,
                #USL_A_dict,
                #ING_B_dict,
                #ALE_B_dict

                #,
                #MLS_dict,
                #CHIN_A_dict,
                #JAP_A_dict,
                #NOR_A_dict
                ]

comps_dict = {'MLS':'MLS',
            'BRA_A': 'BRA A', 
            'BRA_B': 'BRA B', 
            'SUE_A': 'SUE A',
            'CHIN_A': 'CHIN A',
            'JAP_A': 'JAP A',
            'NOR_A': 'NOR A',

            'BEL_A': 'BEL A', 
            'USL_A': 'USL A',
            'ING_B': 'ING B',
            'ALE_B': 'ALE B'
            }

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
import numpy as np
import time
import datetime



for dict in lista_dicts:
       
    comp = dict['comp']
    url = dict['link']
    tamanho_semanas = int(dict['semanas'])
    # Configuração inicial

    # Iniciar o navegador e acessar a URL
    driver = webdriver.Chrome()
    driver.get(url)

    # Lista para armazenar os dados
    data_list = []


    def element_exists(xpath):
        try:
            driver.find_element(By.XPATH, xpath)
            return True
        except:
            return False



    # Função para extrair jogos de uma semana
    def extract_games():
        max_elements = 26
        current_date = None
        for i in range(1, max_elements):
            try:
                date_xpath = f'//*[@id="match-odds"]/div/div[4]/div[{i}]'
                element_text = driver.find_element(By.XPATH, date_xpath).text
                if ("RODADA" in element_text) or ("CONFERÊNCIA" in element_text):
                    current_date = element_text.split(" - ")[0]
                    continue

                match_xpath = f'//*[@id="match-odds"]/div/div[4]/div[{i}]/div/a'
                match_title = driver.find_element(By.XPATH, match_xpath).get_attribute("title")
                home_team, away_team = match_title.split(" - ")

                # Tentar extrair os scores
                try:
                    home_score_xpath = f'//*[@id="match-odds"]/div/div[4]/div[{i}]/div/a/div[3]/div[1]/div/span'
                    home_score = driver.find_element(By.XPATH, home_score_xpath).text
                except:
                    home_score = np.nan

                try:
                    away_score_xpath = f'//*[@id="match-odds"]/div/div[4]/div[{i}]/div/a/div[3]/div[2]/div/span'
                    away_score = driver.find_element(By.XPATH, away_score_xpath).text
                except:
                    away_score = np.nan

                # Abrir o jogo em uma nova aba
                match_link = driver.find_element(By.XPATH, match_xpath).get_attribute("href")
                driver.execute_script(f"window.open('{match_link}', '_blank');")
                driver.switch_to.window(driver.window_handles[1])
                time.sleep(7)

                # Clicar na aba de estatísticas
                stats_tab_xpath = '//*[@id="event-header"]/div[2]/ul[1]/li[2]/a'
                driver.find_element(By.XPATH, stats_tab_xpath).click()
                time.sleep(7)

                # Extrair escanteios
                corners_team1_xpath = '//*[@id="match-stats"]/div/div[2]/div/section[1]/div/div[1]/div[1]/div[1]'  # Adicione o XPath aqui
                corners_team2_xpath = '//*[@id="match-stats"]/div/div[2]/div/section[1]/div/div[1]/div[3]/div[2]'  # Adicione o XPath aqui

                shots_off_target_team1_xpath = '//*[@id="match-stats"]/div/div[2]/div/section[1]/div/div[4]/div[2]/div[1]/span[1]'  # Adicione o XPath aqui
                shots_off_target_team2_xpath = '//*[@id="match-stats"]/div/div[2]/div/section[1]/div/div[4]/div[2]/div[2]/span[1]'  # Adicione o XPath aqui

                shots_on_target_team1_xpath = '//*[@id="match-stats"]/div/div[2]/div/section[1]/div/div[4]/div[2]/div[1]/span[2]'  # Adicione o XPath aqui
                shots_on_target_team2_xpath = '//*[@id="match-stats"]/div/div[2]/div/section[1]/div/div[4]/div[2]/div[2]/span[2]'  # Adicione o XPath aqui

                fouls_team1_xpath = '//*[@id="match-stats"]/div/div[2]/div/section[2]/div/div[3]/div[1]/div[1]'  # Adicione o XPath aqui
                fouls_team2_xpath = '//*[@id="match-stats"]/div/div[2]/div/section[2]/div/div[3]/div[3]/div[2]'  # Adicione o XPath aqui

                red_cards_team1_xpath = '//*[@id="match-stats"]/div/div[2]/div/section[2]/div/div[2]/div[1]/div[1]'  # Adicione o XPath aqui
                red_cards_team2_xpath = '//*[@id="match-stats"]/div/div[2]/div/section[2]/div/div[2]/div[3]/div[2]'  # Adicione o XPath aqui

                # ... [Adicione mais XPaths conforme necessário]

                # Tente extrair cada estatística e, se não conseguir, defina como np.nan
                corners_team1 = driver.find_element(By.XPATH, corners_team1_xpath).text if element_exists(corners_team1_xpath) else np.nan
                corners_team2 = driver.find_element(By.XPATH, corners_team2_xpath).text if element_exists(corners_team2_xpath) else np.nan

                shots_off_target_team1 = driver.find_element(By.XPATH, shots_off_target_team1_xpath).text if element_exists(shots_off_target_team1_xpath) else np.nan
                shots_off_target_team2 = driver.find_element(By.XPATH, shots_off_target_team2_xpath).text if element_exists(shots_off_target_team2_xpath) else np.nan

                shots_on_target_team1 = driver.find_element(By.XPATH, shots_on_target_team1_xpath).text if element_exists(shots_on_target_team1_xpath) else np.nan
                shots_on_target_team2 = driver.find_element(By.XPATH, shots_on_target_team2_xpath).text if element_exists(shots_on_target_team2_xpath) else np.nan

                fouls_team1 = driver.find_element(By.XPATH, fouls_team1_xpath).text if element_exists(fouls_team1_xpath) else np.nan
                fouls_team2 = driver.find_element(By.XPATH, fouls_team2_xpath).text if element_exists(fouls_team2_xpath) else np.nan

                red_cards_team1 = driver.find_element(By.XPATH, red_cards_team1_xpath).text if element_exists(red_cards_team1_xpath) else np.nan
                red_cards_team2 = driver.find_element(By.XPATH, red_cards_team2_xpath).text if element_exists(red_cards_team2_xpath) else np.nan

                # ... [Extraia mais estatísticas conforme necessário]

                data_list.append({
                    'date': current_date,
                    'team1': home_team,
                    'team2': away_team,
                    'score1': home_score,
                    'score2': away_score,
                    'corners_team1': corners_team1,
                    'corners_team2': corners_team2,
                    'shots_off_target_team1': shots_off_target_team1,
                    'shots_off_target_team2': shots_off_target_team2,
                    'shots_on_target_team1': shots_on_target_team1,
                    'shots_on_target_team2': shots_on_target_team2,
                    'fouls_team1': fouls_team1,
                    'fouls_team2': fouls_team2,
                    'red_cards_team1': red_cards_team1,
                    'red_cards_team2': red_cards_team2
                    # ... [Adicione mais campos conforme necessário]
                })

                # Fechar a aba e voltar para a lista de jogos
                driver.close()
                driver.switch_to.window(driver.window_handles[0])

            except Exception as e:
                print(f"Erro ao processar o jogo {i}: {e}")
                continue


    from selenium.common.exceptions import NoSuchElementException

    # Clicar no botão para abrir o dropdown de seleção de semana
    button_xpath = '//*[@id="match-odds"]/div/div[1]/div[1]/button'
    button = driver.find_element(By.XPATH, button_xpath)
    button.click()

    # Esperar o dropdown abrir
    wait = WebDriverWait(driver, 10)
    wait.until(EC.visibility_of_element_located((By.XPATH, '//*[@id="match-odds"]/div/div[1]/div[1]/div')))

    # Iterar pelas semanas e extrair os jogos
    for week in range(tamanho_semanas, 0, -1):
        try:
            week_option_xpath = f'//*[@id="match-odds"]/div/div[1]/div[1]/div/div[{(tamanho_semanas+1)-week}]'
            week_option = driver.find_element(By.XPATH, week_option_xpath)
        except NoSuchElementException:
            # Se a opção da semana não for encontrada, sair do loop e ir para o próximo campeonato
            break

        week_option.click()

        time.sleep(10)

        extract_games()

        # Reabra o dropdown para a próxima iteração
        button.click()
        wait.until(EC.visibility_of_element_located((By.XPATH, '//*[@id="match-odds"]/div/div[1]/div[1]/div')))

    # Fechar o navegador
    driver.quit()


    # Converter a lista de dicionários em um DataFrame
    df = pd.DataFrame(data_list)
    
    df.to_excel(f"dataframe2023_FEIO_{comp}.xlsx")

    df2 = pd.DataFrame()


    def convert_date(date_string):
        if date_string is None:
            return None
        
        month_to_num = {
            "JAN": "01",
            "FEV": "02",
            "MAR": "03",
            "ABR": "04",
            "MAI": "05",
            "JUN": "06",
            "JUL": "07",
            "AGO": "08",
            "SET": "09",
            "OUT": "10",
            "NOV": "11",
            "DEZ": "12"
        }

        try:
            day = date_string.split()[0]
            month = month_to_num[date_string.split()[1].upper()]
            year = datetime.datetime.now().year
            formatted_date = f"{day}/{month}/{year}"
            
            return formatted_date

        except (IndexError, KeyError):
            return None

    df2['team1'] = df['team1']
    df2['team2'] = df['team2']
    df2['team1_goals'] = df['score1'].astype('float64')
    df2['team2_goals'] = df['score2'].astype('float64')
    df2['season'] = int('2023')
    df2['championship'] = comps_dict[comp]


    df2['team1_shots_on_target'] = df['shots_on_target_team1'].astype('float64')
    df2['team1_shots_out'] = df['shots_off_target_team1'].astype('float64')
    df2['team2_shots_on_target'] = df['shots_on_target_team2'].astype('float64')
    df2['team2_shots_out'] = df['shots_off_target_team2'].astype('float64')
    df2['team1_red_cards'] = df['red_cards_team1'].astype('float64')
    df2['team2_red_cards'] = df['red_cards_team2'].astype('float64')
    df2['team1_goals'] = df['score1'].astype('float64')
    df2['team2_goals'] = df['score2'].astype('float64')
    df2['team1_fouls'] = df['fouls_team1'].astype('float64')
    df2['team2_fouls'] = df['fouls_team2'].astype('float64')
    df2['team1_corners'] = df['corners_team1'].astype('float64')
    df2['team2_corners'] = df['corners_team2'].astype('float64')

    df2['team1_total_shots'] = (df2['team1_shots_out'] + df2['team1_shots_on_target']).astype('float64')
    df2['team2_total_shots'] = (df2['team2_shots_out'] + df2['team2_shots_on_target']).astype('float64')


    # Supondo que df é o DataFrame original e df2 é um DataFrame vazio
    if 'date' in df.columns:
        df2['date'] = df['date'].apply(convert_date)
        df2['date'] = pd.to_datetime(df2['date'], errors='coerce', dayfirst=True)

        # Agora todos os objetos são do tipo datetime64[ns] ou NaT (para entradas inválidas)
        cutoff_date = pd.Timestamp(datetime.datetime.now() - datetime.timedelta(days=2))

        # Lista das colunas que queremos verificar por valores faltantes
        cols_to_check = ['team1_goals', 'team2_goals', 'team1_corners', 'team2_corners', 'team1_fouls', 'team2_fouls',
                        'team1_shots_on_target', 'team1_total_shots', 'team1_shots_out', 'team2_shots_on_target', 'team2_total_shots', 'team2_shots_out']

        # Identifica linhas que têm datas inválidas (NaT)
        invalid_date_rows = df2[df2['date'].isna()]

        # Identifica linhas que têm datas anteriores a 5 dias atrás e valores faltantes nas colunas especificadas
        old_date_rows = df2[(df2['date'] < cutoff_date) & df2[cols_to_check].isna().any(axis=1)]

        # Concatenamos os índices das linhas inválidas e das linhas antigas com valores faltantes
        rows_to_drop = pd.concat([invalid_date_rows, old_date_rows])

        # 2. Dropar essas linhas
        df2.drop(rows_to_drop.index, inplace=True)

        df2.sort_values('date', inplace=True)





    df2.to_excel(f"dataframe2023_{comp}.xlsx")
```
