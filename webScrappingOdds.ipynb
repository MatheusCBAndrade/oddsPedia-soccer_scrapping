```python
import datetime

def convert_date(date_string):
    # Dicionário para converter o mês por extenso para o seu número correspondente
    month_to_num = {
        "JAN": "01",
        "FEV": "02",
        "MAR": "03",
        "ABR": "04",
        "MAI": "05",
        "JUN": "06",
        "JUL": "07",
        "AGO": "08",
        "SET": "09",
        "OUT": "10",
        "NOV": "11",
        "DEZ": "12"
    }

    # Extrair o dia e o mês da string original
    day = date_string.split()[0]
    month = month_to_num[date_string.split()[1]]

    # Obter o ano atual
    year = datetime.datetime.now().year

    # Formatar a data
    formatted_date = f"{day}/{month}/{year}"

    return formatted_date





df2['date'] = df['date'].apply(convert_date)

df2['date'] = pd.to_datetime(df2['date'], format='mixed', dayfirst=True)
df2['date'] = df2['date'].dt.date

df2.sort_values('date', inplace=True)


df2['team1'] = df['team1']
df2['team2'] = df['team2']
df2['team1_goals'] = df['score1'].astype('float64')
df2['team2_goals'] = df['score2'].astype('float64')
df2['season'] = int('2023')
df2['championship'] = 'BRA A'


df2['team1_shots_on_target'] = df['shots_on_target_team1'].astype('float64')
df2['team1_shots_out'] = df['shots_off_target_team1'].astype('float64')
df2['team2_shots_on_target'] = df['shots_on_target_team2'].astype('float64')
df2['team2_shots_out'] = df['shots_off_target_team2'].astype('float64')
df2['team1_red_cards'] = df['red_cards_team1'].astype('float64')
df2['team2_red_cards'] = df['red_cards_team2'].astype('float64')
df2['team1_goals'] = df['score1'].astype('float64')
df2['team2_goals'] = df['score2'].astype('float64')
df2['team1_fouls'] = df['fouls_team1'].astype('float64')
df2['team2_fouls'] = df['fouls_team2'].astype('float64')
df2['team1_corners'] = df['corners_team1'].astype('float64')
df2['team2_corners'] = df['corners_team2'].astype('float64')

df2['team1_total_shots'] = (df2['team1_shots_out'] + df2['team1_shots_on_target']).astype('float64')
df2['team2_total_shots'] = df2['team2_shots_out'] + df2['team2_shots_on_target']

#df2 = df2.dropna(subset=['team1_shots_on_target','team1_shots_out'])
```
